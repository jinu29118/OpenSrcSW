```python
#1: 생일 기온 시각화

import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)

max_temp=[]
min_temp=[]
for row in data:
    if row[-1]!=''and row[0].split('-')[1]=='01'and row[0].split('-')[2]=='06':
        max_temp.append(float(row[-1]))
        min_temp.append(float(row[-2]))

plt.title('06-Jan Max & Min temperature(2010~2020)')
plt.plot(max_temp, label='Max')
plt.plot(min_temp, label='Min')

plt.legend()

plt.show()
    
```

```python
#2 : 최고 기온 시각화

import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)

Jan=[]
Aug=[]

for row in data:
    if row[-1]!='':
        if row[0].split('-')[1]=='01':
            Jan.append(float(row[-1]))
        elif row[0].split('-')[1]=='08':
            Aug.append(float(row[-1]))

plt.hist(Jan,bins=100,label='Jan')
plt.hist(Aug,bins=100,label='Aug')

plt.title('Jan & Aug highest temperature in 2010~2020')
plt.legend()
plt.show()
```
