```python
#1 : 서울의 일교차(19040101~20200524)

import csv
f=open('seoul.csv')
data=csv.reader(f)
header=next(data)

#일교차 최대값, 최대인 날짜 기록할 변수 초기화
diff_max=0
diffmax_date=''

#최고기온, 최저기온, 각각의 날짜를 기록할 변수 초기화
max_temp=-999
max_date=''
min_temp=999
min_date=''

# 인덱스 0:날짜, -1: 최고기온 -2:최저기온
for row in data:
    if row[-1]=='':
        row[-1]=-999
    row[-1]=float(row[-1])
    
    if row[-2]=='':
        row[-2]=999
    row[-2]=float(row[-2])
    
    if max_temp<row[-1]:
        max_temp=row[-1]
        max_date=row[0]
    if min_temp>row[-2]:
        min_temp=row[-2]
        min_date=row[0]
    if diff_max<row[-1]-row[-2]:
        diff_max=row[-1]-row[-2]
        diffmax_date=row[0]
f.close()

print('일교차가 가장 큰 날\n{}'.format(diffmax_date))
print('최저기온이 가장 낮은 날\n{}'.format(min_date))
print('최고기온이 가장 높은 날\n{}'.format(max_date))
```
